# Aceleradores descenso de gradiente

```python
import numpy as np
```

```python
def armijo_conditions(X,g,ùõº,f):
    
    c = 0.8

    if f(X-ùõº*g)<f(X)-c*ùõº*np.dot(g, g):
        return True
    return False
```

```python
def backtracking(X,g,f):
    ùúå = 0.9
    ùõº = 100
    
    while True:
        if armijo_conditions(X,g,ùõº,f):
            return ùõº
        ùõº = ùúå*ùõº
```

```python
def gradientN(X,f,n):
    grad = np.zeros(n)
    h = 1e-12
    for i in range(0,n-1):
        e = np.zeros(n)
        e[i] = 1
        grad[i] = (f(X+e*h)-f(X))/h
    
    return grad
```

```python
def rosenbrockN(X):
    sum = 0
    n = len(X)
    
    for i in range(0,n-2):
        sum = sum + (1-X[i])**2 + 100*(X[i+1]-X[i]**2)**2
    return sum
```

```python
def momentum(X,f):
    n = len(X)
    p = np.zeros(n)
    
    max_iter = 50000
    toler = 1e-5
    
    n1 = 0.9
    print("Empezando..")
    
    for i in range(0,max_iter):
        g = gradientN(X,f,n) 
        Œ± = backtracking(X,g,f) #* (-1)
        
        p = n1*p + (g * Œ±)
        
        if(np.linalg.norm(g)<toler):
            break
        print("Iteracion: "+str(i)+" \t||g||: "+str(np.linalg.norm(g)))
        #print("X: "+str(X))
        
        X = X - p
    print("Terminando..")    
    return X
```
