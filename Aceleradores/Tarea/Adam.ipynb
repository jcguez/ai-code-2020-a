# Aceleradores descenso de gradiente

```python
import numpy as np
```

```python
def gradientN(X,f,n):
    grad = np.zeros(n)
    h = 1e-12
    for i in range(0,n-1):
        e = np.zeros(n)
        e[i] = 1
        grad[i] = (f(X+e*h)-f(X))/h
    
    return grad
```

```python
def rosenbrockN(X):
    sum = 0
    n = len(X)
    
    for i in range(0,n-2):
        sum = sum + (1-X[i])**2 + 100*(X[i+1]-X[i]**2)**2
    return sum
```

```python
def backtracking(X,g,f):
    洧랣 = 0.9
    洧띺 = 100
    
    while True:
        if armijo_conditions(X,g,洧띺,f):
            return 洧띺
        洧띺 = 洧랣*洧띺
```

```python
def armijo_conditions(X,g,洧띺,f):
    
    c = 0.8

    if f(X-洧띺*g)<f(X)-c*洧띺*np.dot(g, g):
        return True
    return False
```

### Adam 
Adem치s de tener un tama침o de paso diferente para cada entrada, la penalizaci칩n por los cambios va decayendo. Adem치s utiliza el cuadrado de los gradientes

#### (Adaptive Moment Estimation)

```python
def adam(X,f):
    n = len(X)
    G = np.zeros(n)
    v = np.zeros(n)
    p = np.zeros(n)
    
    e = 1e-8
    max_iter = 50000
    toler = 1e-5
    
    n1 = 0.9
    n2 = 0.99
    
    print("Empezando..")
    
    for i in range(0,max_iter):
        g = gradientN(X,f,n)
        G = g**2
        
        #p = (1-n1)*p + (n1*g)
        p = n1*p + (1-n1)*g
        
        v = n2*v + (1-n2)*G
        
        if(np.linalg.norm(g)<toler):
            break
        print("Iteracion: "+str(i)+" \t||g||: "+str(np.linalg.norm(g)))
        print("X: "+str(X))
        
        풤  = backtracking(X,g,f)
        
        X = X - ((풤 * (p/(1-n1))) / (np.sqrt(v/(1-n2)) + e))
    print("Terminando..")    
    return X
```
